```python
df["Fare_INR"] = df["Fare"].apply(lambda x: x*80)
```

```python
df.head()
```

```python
def create_flag(x):
    if x < 10:
        return "cheap"
    elif 10 <= x < 20:
        return "medium"
    else:
        return "high"
```

```python
df["flag_fare"] = df["Fare"].apply(create_flag)
```
