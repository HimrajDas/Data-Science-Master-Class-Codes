# SMOTE(Synthetic Minority Oversampling Techique)

```python
from sklearn.datasets import make_classification
```

```python
x, y = make_classification(n_samples=1000, n_redundant=0, n_features=2, n_clusters_per_class=1, weights=[0.90], random_state=12)
```

```python
x.shape
```

```python
y.shape
```

```python
x
```

```python
y
```

```python
import pandas as pd

df1 = pd.DataFrame(x, columns=["f1", "f2"])
df2 = pd.DataFrame(y, columns=["target"])
final_df = pd.concat([df1, df2], axis=1)
```

```python
final_df.head()
```

```python
final_df["target"].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.scatter(final_df["f1"], final_df["f2"], c=final_df["target"], cmap="viridis")
plt.title("Scatter Plot of f1 and f2")
plt.colorbar()
plt.grid(True)
```

```python
# !pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# transform the dataset

oversample = SMOTE()
x, y = oversample.fit_resample(final_df[["f1", "f2"]], final_df["target"])
```

```python
x.shape
```

```python
y.shape
```

```python
df1 = pd.DataFrame(x, columns=["f1", "f2"])
df2 = pd.DataFrame(y, columns=["target"])
oversample_df = pd.concat([df1, df2], axis=1)
```

```python
plt.scatter(oversample_df["f1"], oversample_df["f2"], c=oversample_df["target"], s=6)
```
