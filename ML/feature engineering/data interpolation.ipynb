# Data Interpolation

Data interpolation is the process of estimating unknown values within a dataset based on the known values. in Python,
there are various libraries available that can be used for Data Interpolation such as Numpy, SciPy, pandas. Some examples of how
to perform Data Interpolation:

1. Linear Interpolation
2. Cubic Interpolation with SciPy
3. Polynomial Interpolation

## 1. Linear Interpolation

```python
import numpy as np
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)
```

```python
new_x = np.linspace(1, 5, 10)
y_interp = np.interp(new_x, x, y)
```

```python
y_interp
```

```python
plt.scatter(new_x, y_interp)
```

## 2. Cubic Interpolation With SciPy

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 8, 27, 64, 125])
```

```python
from scipy.interpolate import interp1d
```

```python
# create a cubic interpolation function
func = interp1d(x, y, kind="cubic")
```

```python
# interpolate the data
x_new = np.linspace(1, 5, 10)
y_interp = func(x_new)
```

```python
y_interp
```

```python
x_new
```

```python
plt.scatter(x, y, s=18)
plt.grid(True)
```

```python
plt.scatter(x_new, y_interp, s=10)
plt.grid(True)
```

## Polynomial Interpolation

```python
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
```

```python
# interpolate the data using polunomial interpolation
p = np.polyfit(x, y, 2)
```

```python
x_new = np.linspace(1, 5, 10)
y_interp = np.polyval(p, x_new)
```

```python
x_new
```

```python
y_interp
```

```python
plt.scatter(x, y, s=8)
plt.grid(True)
```

```python
plt.scatter(x_new, y_interp, s=14)
plt.grid(True)
```
